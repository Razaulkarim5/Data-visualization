**You have created a dataset `df`. Plot `c vs d` as a scatter plot to recreate the plot below. Try to match the color and size of the data points. You should be able to match the size of the figure as well!**

```python
# Code goes here!
plt.figure(figsize=(7, 3))  # Set figure size
plt.scatter(df['c'], df['d'], c='green', s=30, alpha=0.7)  # Set color and size of points

plt.xlabel('c')
plt.ylabel('d')
plt.show()

```

** Create a histogram of the 'd' column.**

```python
# Code goes here!
plt.figure(figsize=(4, 3))  # Set figure size
df['d'].hist()
plt.grid(False)
```

** Change style sheets to set for 'bmh' and add more bins (e.g. 50) to the above plot.**

```python
# Code goes here!
plt.figure(figsize=(5, 3))  # Set figure size
plt.style.use('bmh')
df['d'].hist(bins=50)
```

**Create a scatter plot for a vs b. How to use the s parameter to indicate the size of the data point based of col d?**<br>

```python
# Code goes here!
plt.figure(figsize=(4, 3))  # Set figure size
plt.scatter(df['a'], df['b'], c='green', s=30, alpha=0.7)  # Set color and size of points

plt.xlabel('b')
plt.ylabel('a')
plt.show()
```

**Recreate the plot below**

```python
# Code goes here!
plt.figure(figsize=(4, 3))
hb = plt.hexbin(df['a'], df['b'], gridsize=30, cmap='coolwarm', mincnt=1)
plt.colorbar(hb, label='Frequency')  # Add a color bar
plt.xlabel('a', fontsize=12)
plt.ylabel('b', fontsize=12)

plt.show()
```
